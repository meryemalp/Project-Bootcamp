```python
df = pd.read_csv("ev_fiyat_tahmini.csv", sep=";")
```

```python
plt.figure(figsize=(15,7))
sp.scatterplot(data=df, x='MetreKare', y='fiyatlar')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
veri = df
```

```python
X = veri[['MetreKare']]
y = veri['fiyatlar']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Metrekare')
plt.ylabel('Fiyat tahmini')
plt.title('Lineer Regresyon')
plt.show()
```

```python
r2_skoru = model.score(X, y)
print("R-kare Skoru:", r2_skoru)
```
